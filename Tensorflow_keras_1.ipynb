### Getting Started withKeras-Sequential Model Creation

```python
import tensorflow as tf
print(tf.__version__)
```

```python
# import libraries

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Softmax
```

### Creating feedforward Model with 3 Layers

```python
model = Sequential([
      Flatten(input_shape=(28,28)),
      Dense(16, activation = "relu"),
      Dense(16, activation = "relu"),
      Dense(10, activation="softmax")
])
```

### Printing Model Summary with wight and summary

```python
model.weights
```

```python
model.summary()
```

### N.B The default activation function of Dense Layer is LINEAR

```python
model = Sequential([
      Flatten(input_shape=(28,28)),
      Dense(16, activation = "relu"),
      Dense(16, activation = "relu"),
      Dense(10),
      Softmax()
])
```

```python
model.summary()
```
